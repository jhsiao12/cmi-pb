## Explore Ontology given

I saw that they had dropped a link to this and I wanted to see if this helps our project in any way. URL link [here](https://www.cmi-pb.org/terminology/owl:AnnotationProperty) 

```python
from rdflib import Graph

g = Graph()
ontology_url = "https://www.cmi-pb.org/myterminology/cmi-pb.owl"
g.parse(ontology_url, format='xml')

query = """
SELECT ?subject ?predicate ?object
WHERE {
  ?subject ?predicate ?object
}
LIMIT 100
"""

results = g.query(query)

for row in results:
    print(f"Subject: {row.subject}, Predicate: {row.predicate}, Object: {row.object}")
```
